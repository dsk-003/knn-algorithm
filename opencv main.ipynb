```python
# importing opencv-python (cv2) Module
import cv2 as cv
# importing matplotlib
import matplotlib.pyplot as plt

# Read the image in Color mode mode
img = cv.imread('F:/PythonML/Python Datasets/Classification Datasets/earth-globe.jpg',1)
# Convert from BGR Color mode to RGB Color mode
RGBimg = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Display image using matpotlib
plt.imshow(RGBimg)
```
