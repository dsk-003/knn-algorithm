```python
X=IrisData.iloc[:,:4]# x is consider as sepal length width y will be species or predict variable i location means index location
X.head()
```

```python
Y=IrisData.iloc[:,-1]#starting from last location

Y.head()
```

```python
#normalixzation ur data preprocee ur value of x and y split ur data
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X=sc_X.fit_transform(X)
X[0:4]
```

```python
from sklearn.model_selection import train_test_split# split into training and testing the data
```

```python
#TRAIN SPLIT DATAT


X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=1,)

Y_test.shape# out of 100% 30% is for testing the data 70% is for go  training .shape 45 row kept for testing
#the data randomly slect
```

```python
#value of k=3
#fit the model predict the class label X_test
knnmodel=KNeighborsClassifier(n_neighbors=3)
knnmodel.fit(X_train,Y_train)
Y_predict1=knnmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score #accuracy of the model 
```

```python
acc=accuracy_score(Y_test,Y_predict1)#these are value for preducting the data i e 97%
```
